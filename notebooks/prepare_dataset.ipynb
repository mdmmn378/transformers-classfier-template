## Import Modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import dask.dataframe as dd
import glob
import matplotlib.pyplot as plt
sns.set_palette("flare")
%matplotlib inline
```

## Load Datasets

```python
def read_topic(paths):
    tmp = pd.concat([pd.read_excel(i) for i in paths])
    return tmp.get(["body", "cluster_id"])
```

```python
mens_health = glob.glob("../datasets/MLDataset/Used_Automation/mens_health/**.xlsx")
menstruation = glob.glob("../datasets/MLDataset/Used_Automation/mensturation/**.xlsx")
others = glob.glob("../datasets/MLDataset/Used_Automation/Others/**.xlsx")
pain = glob.glob("../datasets/MLDataset/Used_Automation/pain/**.xlsx")
pregnancy = glob.glob("../datasets/MLDataset/Used_Automation/pregnancy/**.xlsx")
```

```python
df_mens_health = read_topic(mens_health)
df_menstruation = read_topic(menstruation)
df_others = read_topic(others)
df_pain = read_topic(pain)
df_pregnancy = read_topic(pregnancy)
```

## EDA

```python
def plot_label_distribution(dataset):
    plt.figure(figsize=(20, 20))
    bins = dataset["cluster_id"].nunique()
    sns.displot(dataset["cluster_id"], bins=bins)
    plt.show()
```

```python
plot_label_distribution(df_mens_health)
```

```python
df_others.cluster_id.value_counts()
```

```python
plot_label_distribution(df_others)
```

```python
plot_label_distribution(df_pain)
```

```python
plot_label_distribution(df_pregnancy)
```
